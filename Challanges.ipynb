### Using groupby, plot the number of medals awarded at each of the olympics games throughout history

```python
edi = oo.groupby('Edition').size().plot()
edi
```

### Create a list showing the total number of medals won by each country over the history of the Olympics. For each countrym include the year of the first most recent olympic medal wins.

```python
oo.groupby('NOC').agg({'Edition':['min','max','count']})
```

### Plot the number of gold medals won by the US male and female athletes  throughout the history of the olympics ?

```python
gm = oo[(oo.Medal == 'Gold') & (oo.NOC == 'USA')]
gm
```

```python
gm.groupby(['Gender', 'Edition']).size().unstack('Gender', fill_value=0).plot()
```

### Plot the five athletes who have won the most gold medals over the history of the olympics. when there is a tie, consider the number of silver medals, and the bronze medals.

```python
#my solution
agm = oo[oo.Medal == 'Gold']
agm.Athlete.value_counts().head(5)
```

```python
#answer
g = oo.groupby(['Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']].head().plot(kind='bar')
```
