#### Which country has won the most men's gold medals in singles badminton over the years? Sort the results alphabetically by the player's names.

```python
#badminton = oo[['NOC','Sport', 'Gender','Medal', 'Edition', 'Athlete']]
#badminton

#Solution
gbm = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Men') & (oo.Sport == 'Badminton')]
gbm.sort_values(by='Athlete')
```

#### Which three countries have won the most medals in recent years?

```python
#recent = oo[['NOC', 'Medal', 'Gender', 'Event', 'Athlete']]
#recent

#Solution
rec = oo[oo.Edition >= 1984]
rec.NOC.value_counts().head(3)
```

### Display the male gold medal winners for the 100 meters track and field sprint event over the years. List the results starting with the most recent. Show the Olympic city, edition, athlete, and the country they represent.

```python
#[(recent.Gender == 'Male') & (recent.Event == '100m track')]

#Solution
gmh = oo[(oo.Gender == 'Men') & (oo.Medal  == 'Gold') & (oo.Event == '100m')]
#most recent first
gmh.sort_values('Edition', ascending=False)[['City', 'Edition', 'Athlete', 'NOC']]
```

## Stack() and unstack()

```python
mw = oo[(oo.Edition == 2008) & ( (oo.Event == '100m') | (oo.Event == '200m') )] 
mw
```

```python
g = mw.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
g
```

```python
df = g.unstack(['Discipline', 'Event'])
df
```

```python
df.stack()
```

```python
df.stack('Event')
```

```python
df.unstack()
```

```python
df.unstack('Gender')
```

```python
lo = oo[oo.Edition == 2008]
lo
```

```python
g = lo.groupby(['NOC','Medal']).size().unstack('Medal', fill_value=0)
g = g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold','Silver','Bronze']]
g
```

```python
import seaborn           as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
g = g.transpose()
g
```

```python
plt.figure(figsize=(16,5))
sns.heatmap(g)
```
